# Section 4: Merging the dataframes

```python
# convert the timestamp into hourly data as per the problem statement
# problem statement: Can we accurately predict the stock levels of products based on sales data and sensor data on an hourly basis in order to more intelligently procure products from our suppliers?

sales_df['timestamp'] = sales_df['timestamp'].dt.strftime('%Y-%m-%d %H:00:00')
stock_df['timestamp'] = stock_df['timestamp'].dt.strftime('%Y-%m-%d %H:00:00')
temp_df['timestamp'] = temp_df['timestamp'].dt.strftime('%Y-%m-%d %H:00:00')
```

```python
sales_df.head()
```

```python
stock_df.head()
```

```python
temp_df.head()
```

```python
sales_agg = sales_df.groupby(['timestamp', 'product_id']).agg({'quantity': 'sum'}).reset_index()
sales_agg.head()
```

```python
# average stock pct at any given time
stock_agg = stock_df.groupby(['timestamp', 'product_id']).agg({'estimated_stock_pct': 'mean'}).reset_index()
stock_agg.head()
```

```python
# average temperature at a certain hour
temp_agg = temp_df.groupby(['timestamp']).agg({'temperature': 'mean'}).reset_index()
temp_agg.head()
```

```python
merged_df = stock_agg.merge(
    sales_agg,
    how='left',
    on=['timestamp', 'product_id']
)

merged_df.head()
```

```python
merged_df = merged_df.merge(
    temp_agg,
    how='left',
    on='timestamp'
)

merged_df.head()
```

```python
merged_df.info()
```

```python
merged_df.fillna(0, inplace=True)
```

```python
merged_df.head()
```

```python
sales_df.head()
```

```python
product_categories = sales_df[['product_id', 'category', 'unit_price', 'customer_type', 'payment_type', 'total']]
product_categories = product_categories.drop_duplicates()
```

```python
product_categories.head()
```

```python
merged_df = merged_df.merge(
    product_categories,
    how='left',
    on='product_id'
)

merged_df.head()
```

# Section 5: Feature Engineering
Problem statement: Can we accurately predict the stock levels of products based on sales data and sensor data on an hourly basis in order to more intelligently procure products from our suppliers?

```python
# want to predict estimated_stock_pct
# one-hot encoding on category

from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown='ignore', sparse_output=False).set_output(transform='pandas')
```

```python
category_enc = enc.fit_transform(merged_df[['category']])
merged_df = merged_df.merge(
    category_enc,
    left_index=True,
    right_index=True
)

merged_df.head()
```

```python
merged_df.drop(
    columns=['customer_type', 'payment_type', 'category', 'total'],
    inplace=True
)
```

```python
merged_df.info()
```

```python
merged_df['timestamp'] = pd.to_datetime(merged_df['timestamp'])
merged_df['timestamp_day_of_month'] = merged_df['timestamp'].dt.day
merged_df['timestamp_day_of_week'] = merged_df['timestamp'].dt.dayofweek
merged_df['timestamp_hour'] = merged_df['timestamp'].dt.hour
merged_df.drop(columns=['timestamp'], inplace=True)
merged_df.head()
```

# Section 6: Modelling

```python
# use random forest regressor to model the data
# k fold cross validation
# pick out most important features

from sklearn.model_selection import KFold, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler

# regr = RandomForestRegressor(random_state=42)
# k_folds = KFold(n_splits=5)
# scaler = StandardScaler()

# X = merged_df.drop(columns=['estimated_stock_pct', 'product_id'])
# scaler.fit(X)

# y = merged_df[['estimated_stock_pct']]

# X_train = scaler.transform(X)

# scores = cross_val_score(regr, X_train, y.values.ravel(), cv=k_folds, scoring='neg_mean_absolute_error')
# print("Cross Validation Scores: ", scores)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

k = 10
accuracy = []

X = merged_df.drop(columns=['estimated_stock_pct', 'product_id'])
y = merged_df['estimated_stock_pct']

for fold in range(0, k):
  regr = RandomForestRegressor(random_state=42)
  scaler = StandardScaler()

  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

  scaler.fit(X_train)
  X_train = scaler.transform(X_train)
  X_test = scaler.transform(X_test)

  model = regr.fit(X_train, y_train)
  y_pred = model.predict(X_test)

  mae = mean_absolute_error(y_true=y_test, y_pred=y_pred)
  accuracy.append(mae)

  print(f"Fold {fold + 1}: MAE = {mae:.3f}")

print(f'Average MAE = {sum(accuracy)/len(accuracy):.2f}')
```

```python
regr.feature_importances_
```

```python
from matplotlib import pyplot as plt

# Get features and turn it into a Series
importances = pd.Series(regr.feature_importances_, index=X.columns)
importances.sort_values(ascending=True, inplace=True)

fig, ax = plt.subplots(figsize =(16, 9))

# grid lines
ax.grid(
    color ='grey',
    linestyle ='-.',
    linewidth = 0.5,
    alpha = 0.2
)

# Horizontal bar plot
ax.barh(importances.index, importances)

# Set title
ax.set_title('Relative feature importance')

plt.show()
```

# Conclusion
We can see that the most important features for the random forest model are: unit_price, temperature, timestamp_hour, quantity, timestamp_day_of_month, timestamp_day_of_week.

We can drop the other columns to achieve greater efficiency of our model.

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import GridSearchCV

K = 10
ACCURACY = []
SPLIT = 0.75

def load_data(path):
  # Function takes in a string for the filepath
  df = pd.read_csv(path)

  # Remove unamed column, then returns a Pandas DataFrame
  df.drop(columns=['Unamed: 0'], inplace=True, errors='ignore')
  return df

def create_target_and_predictors(df, target='estimated_stock_pct'):
  # Takes in DataFrame, then drops 'estimated_stock_pct' column, and assigns X and y variables accordingly
  X = df.drop(target)
  y = df[target]
  return X, y

def train_algorithm_with_cv(X, y):
  # function takes in predictor and target variables, trains the model with cross validation, then prints average MAE
  for fold in range(0, K):
    # instantiate the Random Forest Regressor and Standard Scaler
    regr = RandomForestRegressor(random_state=42)
    scaler = StandardScaler()

    # split data randomly for testing
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=SPLIT, random_state=42)

    # scale the data
    scaler.fit(X_train)
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)

    # model training and predicting
    model = regr.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # get mean absolute error, then append to ACCURACY list
    mae = mean_absolute_error(y_true=y_test, y_pred=y_pred)
    ACCURACY.append(mae)

    print(f"Fold {fold + 1}: MAE = {mae:.3f}")

  # compute average MAE across all folds
  print(f'Average MAE = {sum(ACCURACY)/len(ACCURACY):.2f}')
```
